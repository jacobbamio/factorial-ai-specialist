```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import random
import requests
```

## Step By Step Process

```python
topics = [
    "communication", 
    "interfaces design", 
    "backend engineer", 
    "frontend engineer", 
    "team leading", 
    "time management", 
    "meetings management"
]
```

```python
mapped_links = {}
for topic in topics:
    mapped_links[topic] = {}
    mapped_links[topic]['links'] = []
    
    chrome_options = Options()
    chrome_options.add_argument("--disable-search-engine-choice-screen")
    driver = webdriver.Chrome(chrome_options)

    url = f"https://www.udemy.com/courses/search/?src=ukw&q={topic}"
    driver.get(url)
    time.sleep(5)
    last_height = driver.execute_script("return document.body.scrollHeight")
    for i in range(15): 
        
        # Simulate random scroll down
        scroll_down = random.randint(200, 400)  
        driver.execute_script(f"window.scrollBy(0, {scroll_down});")
        time.sleep(random.uniform(1, 3)) # Dynamic wait to make it more realistic
        
        # Fake some scrolls up each 3 scrolls down
        if i % 3 == 0:
            scroll_up = random.randint(150, 300)
            driver.execute_script(f"window.scrollBy(0, -{scroll_up});")
            time.sleep(random.uniform(1, 2)) # Dynamic wait to make it more realistic
   
    time.sleep(5)

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    for l in soup.find_all('link', {'data-testid': 'view-prefetch'}):
        mapped_links[topic]['links'].append(f'https://www.udemy.com{l["href"]}')
    driver.quit()
```

```python
mapped_links
```

```python
trainings_to_db = []
for topic in mapped_links:
    for link in mapped_links[topic]['links']:
        print(topic, link)
        chrome_options = Options()
        chrome_options.add_argument("--disable-search-engine-choice-screen")
        driver = webdriver.Chrome(chrome_options)
        driver.get(link)
        time.sleep(5)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        title = soup.find(class_='ud-heading-xl clp-lead__title clp-lead__title--small').text
        subtitle = soup.find(class_='ud-text-md clp-lead__headline').text
        key_learnings = []

        for wywl in soup.find_all(class_='what-you-will-learn--objective-item--VZFww'):
            key_learnings.append(wywl.text)

        course_dict = {"topic" : topic,
                       "title" : title,
                       "subtitle" : subtitle,
                       "key_learnings" : key_learnings,
                       "link" : link}
        trainings_to_db.append(course_dict)
        print(course_dict)
        driver.quit()
        time.sleep(2)
```

```python
import json

with open("trainings.json", 'w') as fout:
    json.dump({"trainings": trainings_to_db}, fout, indent=4)
```

## Upload to Cosmos DB Database

```python
import json
import uuid
from azure.cosmos import exceptions, CosmosClient, PartitionKey
import os 
from dotenv import load_dotenv

load_dotenv()

endpoint = os.getenv("UDEMY_TRAININGS_ENDPOINT")
key = os.getenv("UDEMY_TRAININGS_PRIMARY_KEY")
client = CosmosClient(endpoint, key)

database_name = os.getenv("UDEMY_TRAININGS_DATABASE")
container_name = os.getenv("UDEMY_TRAININGS_CONTAINER")
database = client.create_database_if_not_exists(id=database_name)
container = database.create_container_if_not_exists(
    id=container_name,
    partition_key=PartitionKey(path="/topic"),
    offer_throughput=400
)

with open('trainings.json') as f:
    data = json.load(f)

for training in data['trainings']:
    training['id'] = str(uuid.uuid4())
    container.create_item(body=training)
```
